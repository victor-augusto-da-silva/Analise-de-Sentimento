Classificação Inteligente de Feedback (NLP & ML)Este projeto foi desenvolvido para automatizar a análise de sentimentos e a categorização temática de feedbacks de alunos (NPS/CSAT) na Cogna Educação. A solução utiliza uma abordagem híbrida, combinando regras heurísticas especializadas com modelos de Machine Learning (Scikit-Learn) para garantir alta precisão e escalabilidade. 🚀 FuncionalidadesClassificação de Sentimento: Identifica se o comentário é Promotor, Neutro ou Detrator. Categorização de Tópicos: Classifica automaticamente o feedback em temas como: Professor/Conteúdo, Erro - BUG, Áudio, Qualidade do Vídeo, Edição, entre outros. Abordagem Híbrida: Utiliza um motor de regras para termos críticos e um modelo de Regressão Logística para textos mais complexos. Pipeline de Dados: Processa múltiplos arquivos .parquet simultaneamente, com sistema de backup e rollback automático para segurança dos dados. Exportação para Validação: Scripts auxiliares para converter os resultados em Excel, facilitando a auditoria humana.🛠️ Stack TécnicaLinguagem: PythonProcessamento de Dados: Pandas, NumPy, GlobMachine Learning: Scikit-Learn (TF-IDF Vectorizer, Logistic Regression, Pipelines) Ambiente: AWS (Cloud Storage e Processamento) NLP: Regular Expressions (Re), Unicodedata (Normalização de texto)🏗️ Arquitetura do CódigoPré-processamento: Limpeza de ruídos, remoção de acentos e normalização de strings.Motor de Regras: Identificação de palavras-chave críticas para classificação imediata.Modelagem Preditiva: - Vetorização de texto usando TF-IDF (ngram range 1-2).Treinamento de Regressão Logística com balanceamento de classes (class_weight='balanced'). Camada de Combinação: Lógica de decisão que prioriza regras de negócio e utiliza o ML para preencher lacunas.📂 Como ExecutarCertifique-se de ter os arquivos .parquet na pasta de origem configurada.Instale as dependências:Bashpip install pandas scikit-learn pyarrow
Execute o script principal:Bashpython nps_classification_pipeline.py
📈 Resultados AlcançadosA implementação deste pipeline reduziu drasticamente o tempo de análise manual de feedbacks, permitindo que a equipe de BI foque em insights estratégicos em vez de rotulagem manual de dados.
